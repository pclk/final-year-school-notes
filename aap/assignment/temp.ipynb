```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data preparation
metrics = ["MAE", "MSE", "RMSE", "R2", "RMSLE", "MAPE"]
clustering_values = [22329.7560, 1631662912.7684, 40135.9446, 0.5502, 0.8158, 1.3685]
word2vec_values = [22630.0387, 1675490505.0942, 40643.7717, 0.5389, 0.7982, 1.2626]


# Create DataFrame
df = pd.DataFrame(
    {
        "Metric": metrics,
        "Clustering": clustering_values,
        "Word2Vec": word2vec_values,
    }
)

# Set style and custom color palette
plt.style.use("seaborn")
custom_palette = [
    "#415866",
    "#4f2422",
    "#8b786e",
    "#F57229",
]  # Steel blue, Dark red, Taupe
sns.set_palette(custom_palette)

# Set figure background color to white for better contrast
plt.rcParams["figure.facecolor"] = "white"
plt.rcParams["axes.facecolor"] = "white"

# Create high resolution figure with multiple subplots
plt.rcParams["figure.dpi"] = 600  # Set high DPI for display
plt.rcParams["figure.figsize"] = [16, 20]  # Larger figure size
plt.rcParams["font.size"] = 20  # Larger default font size
plt.rcParams["axes.linewidth"] = 2  # Thicker axes lines
plt.rcParams["axes.labelsize"] = 20  # Larger axis labels
plt.rcParams["xtick.labelsize"] = 20  # Larger tick labels
plt.rcParams["ytick.labelsize"] = 20
plt.rcParams["legend.fontsize"] = 20
plt.rcParams["figure.titlesize"] = 20

# Create figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 8))

# Plot 1: Error Metrics (MAE, RMSE)
metrics_to_plot1 = ["MAE", "RMSE"]
df_main = df[df["Metric"].isin(metrics_to_plot1)].melt(
    id_vars=["Metric"], var_name="Method", value_name="Value"
)

sns.barplot(
    data=df_main,
    x="Metric",
    y="Value",
    hue="Method",
    ax=ax1,
    palette=custom_palette[:3],
)
ax1.set_title("Error Metrics Comparison", pad=20, fontsize=20, fontweight="bold")
ax1.set_ylabel("Value (lower is better)")
ax1.grid(True, alpha=0.3)
# Add value labels on the bars
for container in ax1.containers:
    ax1.bar_label(container, fmt="%.2f", padding=3)

# Plot 2: Performance Metrics (R2, RMSLE, MAPE)
metrics_to_plot2 = ["R2", "RMSLE", "MAPE"]
df_secondary = df[df["Metric"].isin(metrics_to_plot2)].melt(
    id_vars=["Metric"], var_name="Method", value_name="Value"
)

sns.barplot(
    data=df_secondary,
    x="Metric",
    y="Value",
    hue="Method",
    ax=ax2,
    palette=custom_palette[:3],
)
ax2.set_title("Performance Metrics Comparison", pad=20, fontsize=20, fontweight="bold")
ax2.set_ylabel("Value (Lower better except R²)")
ax2.grid(True, alpha=0.3)
# Add value labels on the bars
for container in ax2.containers:
    ax2.bar_label(container, fmt="%.4f", padding=3)

# Enhance grid and spines for both plots
for ax in [ax1, ax2]:
    ax.grid(True, linestyle="--", alpha=0.7, linewidth=1.5)
    ax.spines["top"].set_linewidth(2)
    ax.spines["right"].set_linewidth(2)
    ax.spines["bottom"].set_linewidth(2)
    ax.spines["left"].set_linewidth(2)
    ax.set_facecolor("white")
    # Rotate x-axis labels for better readability
    ax.tick_params(axis="x", rotation=45)

# Add overall title
fig.suptitle("Model Performance Metrics", y=1.05, fontsize=24, fontweight="bold")


# Enhance grid and spines
for ax in axes:
    ax.grid(True, linestyle="--", alpha=0.7, linewidth=1.5)
    ax.spines["top"].set_linewidth(2)
    ax.spines["right"].set_linewidth(2)
    ax.spines["bottom"].set_linewidth(2)
    ax.spines["left"].set_linewidth(2)
    ax.set_facecolor("white")

# Adjust layout
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
plt.style.use("seaborn")
plt.rcParams["figure.facecolor"] = "white"
plt.rcParams["axes.facecolor"] = "white"

# Create figure
fig, ax = plt.subplots(figsize=(10, 5))

# Data
models = ["Our Model", "Commercial LLM"]
features = ["Accuracy", "Speed", "Cost", "Ownership"]
our_model = [1, 1, 1, 1]  # 1 represents advantage
commercial_llm = [-1, -1, -1, -1]  # -1 represents disadvantage

# Create positions for bars
x = np.arange(len(features))
width = 0.35

# Create bars
ax.bar(
    x - width / 2, our_model, width, label="Our Model", color="#1a8c3c", alpha=0.7
)  # Darker green
ax.bar(
    x + width / 2,
    commercial_llm,
    width,
    label="Commercial LLM",
    color="#b71c1c",  # Darker red
    alpha=0.7,
)

# Customize plot
ax.set_title("Why Use Our Model?", fontsize=16, fontweight="bold")
ax.set_xticks(x)
ax.set_xticklabels(features, fontsize=12)
ax.legend(fontsize=12)

# Remove y-axis numbers and add horizontal line at y=0
ax.set_yticks([])
ax.axhline(y=0, color="black", linestyle="-", linewidth=0.5)

# Add annotations
annotations_our = [
    "30% reduction in RMSE\n23% increase in R²",
    "0.03 seconds",
    "256MB Model size",
    "Full ownership",
]

annotations_commercial = [
    "Less accurate",
    "~1 second",
    "Multiple GBs",
    "Provider owned",
]


# Enhance grid and spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

# Set y-axis limits to be tighter
ax.set_ylim(-1.2, 1.6)

# Adjust annotation positions to be closer to bars
for i in range(len(features)):
    # Our model annotations (above bars)
    ax.text(
        i - width / 2,
        1.05,  # Moved closer to bars
        annotations_our[i],
        ha="center",
        va="bottom",
        fontsize=10,
        color="#1a8c3c",  # Darker green
    )
    # Commercial LLM annotations (below bars)
    ax.text(
        i + width / 2,
        -1.05,  # Moved closer to bars
        annotations_commercial[i],
        ha="center",
        va="top",
        fontsize=10,
        color="#e74c3c",
    )

# Add subtle grid
ax.grid(True, axis="x", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```
